Python module to extract from spaceranger outputs the matrix of transcription (transcriptome for each ST spot) of a sample/many samples. Give in output an anndata object. Please note that here the transcription data are normalized, Log1p transformed, batch corrected and at the end only the most 350 variables genes are kept for further analyses (see kernel "Normalising and integrating gene expression data", done with scanpy). Also a QC analysis was perform in this script.

#### Import dependencies

```python
import os
import csv
import numpy as np
import pandas as pd
from sklearn.preprocessing import minmax_scale
import csv
import gzip
import scipy.io
import scanpy as sc
import anndata as ad
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
import sys
from scipy.sparse import csr_matrix
```

# Input

Annotation csv files + directory containing transcriptions matrix (see variable "matrix_dir" in Loading per-barcode gene expression data)

# Output

An anndata object with : in adata.X the matrix of transcription (rows=barcode+the sample ID, columns=genes, so each case of the matrix represent the count of RNA for a gene on a precise ST spot), in adata.obs the sample ID, the pathologist annotation, and in index the barcode+the sample ID.
The goal of this anndata object is to use it to perform some machine learning. QC analysis results are also present in output.

```python
output_path="/disk2/user/cormey/outputs/Transcription_objects/034_RJ_QC"
```

# Loading per-barcode gene expression data

```python
# If your SpaceRanger output doesn't include 3 files under 'filtered_feature_bc_matrix' directory, please change the input files 

samples=["V10F03-034_A","V10F03-034_B","V10F03-034_C","V10F03-034_D"]

# Iterate over samples
matrix_data = pd.DataFrame()
for sample in samples:

    matrix_dir = "/disk2/user/cda/SpatialTranscriptomics/processed-data/SpaceRanger-output_v1.0.0-rerun2023/{}/outs/filtered_feature_bc_matrix".format(sample)
    
    # Read in MEX format matrix as table
    mat_filtered = scipy.io.mmread(os.path.join(matrix_dir, "matrix.mtx.gz"))
    
    # list of transcript ids, e.g. 'ENSG00000187634'
    features_path = os.path.join(matrix_dir, "features.tsv.gz")
    feature_ids = [row[0] for row in csv.reader(gzip.open(features_path, mode="rt"), delimiter="\t")]

    # List of barcodes, e.g. 'AAACATACAAAACG-1'
    barcodes_path = os.path.join(matrix_dir, "barcodes.tsv.gz")
    barcodes = [row[0] for row in csv.reader(gzip.open(barcodes_path, mode="rt"), delimiter="\t")]

    # Transform table to pandas dataframe, label rows and columns and transpose
    matrix = pd.DataFrame.sparse.from_spmatrix(mat_filtered)
    matrix.columns = barcodes
    matrix.index = feature_ids
    matrix = matrix.transpose()

    # Enforce there are all (1) on-tissue spots (2) for which annotation in available (by comparing number of image on tissue with number of row in the matrix)
    sample_id=sample.split("-")[1]
    sample_id1=sample_id.split("_")[0]
    sample_id2=sample_id.split("_")[1]
    directory="/disk2/user/cormey/outputs/Tilted_images/{}_RJ.{}1".format(sample_id1,sample_id2)
    files = [f for f in os.listdir(directory)]
    if len(files) != matrix.shape[0]:
        sys.exit(f"{sample} gene matrix do not have all required barcodes! {len(files)},{matrix.shape[0]}")

    # Add sample identifier and merge to one dataframe
    matrix.index = matrix.index.astype(str) + "#" + sample
    matrix_data = pd.concat([matrix_data, matrix]) 
```

# Observe caracteristics of each cell type

```python
#Creation of a dictionnary which associates a barcode and an annotation
barcode_annot={}
for file in os.listdir(annot_file):
    full_path = os.path.join(annot_file, file) #find the path of the annotation file for each sample 
    dataframe=pd.read_csv(full_path) 
    S_id=file.split("_")[0:2] #get sample id
    S_id="_".join(S_id)
    for index,ligne in dataframe.iterrows():
        key=str(ligne.iloc[0])+'#'+str(S_id) 
        barcode_annot[key]=ligne.iloc[1] #add in a dictionnary the sampleid+barcode in key and the annotation in value
```

```python
#Get the pathologist's annotation for each row (=each spot), stock the row mean (=mean reads value) in a dataframe for each cell type
df_cara=pd.DataFrame()
# Browse matrix_data
for index, row in matrix_data.iterrows():
    BC = index
    annotation = barcode_annot.get(BC)
    if pd.isna(annotation):
        annotation="Mixed"
    mean_value = row.mean()  # Calculate the mean value of reads for the current row
    
    # If this annotation column doesn't exist, create a new one
    if annotation not in df_cara.columns:
        df_cara[annotation] = pd.Series(dtype=float)
    
    # Add the mean value to the good column in df_cara
    df_cara.at[index, annotation] = mean_value

# Fill missing values with NaN
df_cara = df_cara.fillna(value=pd.NA)

print(df_cara)
```

```python
# Calculer la moyenne de chaque colonne
column_means = df_cara.mean()

# Afficher les moyennes
print(column_means)
```

# Add pathologist's annotations corresponding to each row (=each spot) to adata.obs 

```python
#Creation of a dictionnary which associates a barcode and an annotation
barcode_annot={}
for file in os.listdir(annot_file):
    full_path = os.path.join(annot_file, file)
    dataframe=pd.read_csv(full_path)
    S_id=file.split("_")[0:2]
    S_id="_".join(S_id)
    for index,ligne in dataframe.iterrows():
        key=str(ligne.iloc[0])+'#'+str(S_id)
        barcode_annot[key]=ligne.iloc[1]
```

```python
#Get the pathologist's annotation for each row (=each spot), stock them in a list, and then add the list as a column in adata.obs
liste_annot=[]
for ligne in adata.obs.index:
    BC=ligne
    if pd.isna(barcode_annot[BC]): #Replace every missing annotations by Mixed
        liste_annot.append("Mixed")
    else:
        liste_annot.append(barcode_annot[BC])
adata.obs['annotation'] = liste_annot
```

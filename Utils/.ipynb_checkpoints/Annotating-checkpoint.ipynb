# Output

a csv file with path of every tiles and their corresponding annotations

```python
n_img=dpath.split("/")[-1]
output_path="/disk2/user/cormey/outputs/outputs_annotated_tiles/{}.csv".format(n_img) #creation of the path to save the outputs
output_path
```

# Function

Creation of a dictionnary which associate the barcode and the path of each tiles

```python
dico={}
for files in os.listdir(dpath): #browse the directory of tiles
    files_path = os.path.join(dpath, files) #rebuild the entire path of each file
    if os.path.isfile(files_path): #check if it's a file
        barcode=files_path.split("_")[5] #4 for non-big
        dico[barcode]=files_path
        
        
```

```python
dico
```

Creation of a dictionnary which associates a barcode and an annotation

```python
barcode_annot={}
dataframe=pd.read_csv(annot)
for ligne in dataframe.iterrows():
    barcode_annot[ligne[1][0]]=ligne[1][1]
```

```python
barcode_annot
```

Creation of a dictionnary which associate a tile's path to an annotations

```python
path_annot={}
for barcode in dico:
    if pd.isna(barcode_annot[barcode]): #Replace every missing annotations by Mixed
        barcode_annot[barcode]="Mixed"
    path_annot[dico[barcode]]=barcode_annot[barcode]
```

```python
path_annot
```

Creation of a csv file with the results

```python
df = pd.DataFrame.from_dict(path_annot, orient='index', columns=['Annotation'])
df.to_csv(output_path, index_label='Path')
```

# Presentation of data

Creation of a dictionnary which count every apparition of each annotation

```python
dico_count={}
for ligne in path_annot.values():
    if ligne in dico_count:
        dico_count[ligne]+=1
    else:
        dico_count[ligne]=1
```

```python
dico_count
```

Creation of an histogram

```python
# Extraction of keys and values
labels = list(dico_count.keys())
values = list(dico_count.values())

# Creation of the histogram
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color='skyblue')

# Ajout de titres et de labels
plt.xlabel('Cell type')
plt.ylabel('Counts')
plt.title('Representation of the number of tiles belonging to each cell type')

# Affichage de l'histogramme
plt.show()
```
